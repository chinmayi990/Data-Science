```python
                                         # Project 2: Data Exploration on Iris Dataset
#Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load dataset
df = pd.read_csv("iris.csv")

# Step 3: Explore dataset structure
print("First 5 Rows:")
print(df.head())

print("\nDataset Info:")
print(df.info())

print("\nStatistical Summary:")
print(df.describe())

# Step 4: Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())

#Data Visualization
#1.Histograms for feature distributions
df.hist(figsize=(10, 6), edgecolor='black')
plt.suptitle("Feature Distributions (Histograms)", fontsize=14)
plt.show()

#2. Pairplot for feature relationships (colored by species)
sns.pairplot(df, hue="species", diag_kind="hist")
plt.suptitle("Pairplot of Features by Species", y=1.02, fontsize=14)
plt.show()

#3.Boxplots to check spread and outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, orient="h")
plt.title("Boxplot of All Features", fontsize=14)
plt.show()

#4. Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap", fontsize=14)
plt.show()
```
